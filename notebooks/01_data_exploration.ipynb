# Data Exploration

In this notebook, we will explore the video data collected from the concert. The goal is to understand the structure of the data, including the total view and close-up view videos, and to perform some initial analyses.

```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Define paths to the video files
total_view_path = 'path/to/total_view_video.mp4'
close_up_view_path = 'path/to/close_up_view_video.mp4'

# Function to load and display a video frame
def load_and_display_frame(video_path, frame_number):
    cap = cv2.VideoCapture(video_path)
    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_number)
    ret, frame = cap.read()
    if ret:
        plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.show()
    cap.release()

# Display a frame from the total view video
load_and_display_frame(total_view_path, 100)

# Display a frame from the close-up view video
load_and_display_frame(close_up_view_path, 100)
```

## Video Metadata

Next, we will extract and display some metadata from the videos, such as frame count, frame rate, and duration.

```python
def get_video_metadata(video_path):
    cap = cv2.VideoCapture(video_path)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    duration = frame_count / fps
    cap.release()
    return frame_count, fps, duration

total_view_metadata = get_video_metadata(total_view_path)
close_up_view_metadata = get_video_metadata(close_up_view_path)

print("Total View Video: ")
print(f" - Frame Count: {total_view_metadata[0]}")
print(f" - Frame Rate: {total_view_metadata[1]}")
print(f" - Duration: {total_view_metadata[2]} seconds\n")
print("Close-Up View Video: ")
print(f" - Frame Count: {close_up_view_metadata[0]}")
print(f" - Frame Rate: {close_up_view_metadata[1]}")
print(f" - Duration: {close_up_view_metadata[2]} seconds")
```
